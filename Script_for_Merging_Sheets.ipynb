```python
import pandas as pd

# Path to the Excel file
file_path = "/content/MERGED LINELIST 2008-2023_deidentified.xlsx"

# Load the Excel file and get sheet names
xls = pd.ExcelFile(file_path)
sheet_names = xls.sheet_names  # Get the actual sheet names

# Print the actual sheet names to verify
print("Sheet names in the file:", sheet_names)

# List to store DataFrames
df_list = []

# Loop through the actual sheet names and read each one
for sheet_name in sheet_names:
    print(f"Reading sheet: {sheet_name}")
    data = pd.read_excel(file_path, sheet_name=sheet_name)
    df_list.append(data)

# Concatenate all DataFrames
df = pd.concat(df_list, ignore_index=True)

# Save the final DataFrame to CSV
final_csv_path = "/content/final.csv"
df.to_csv(final_csv_path, index=False)

print(f"Data successfully saved to {final_csv_path}")
```
